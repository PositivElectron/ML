```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1,noise=80,random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
# Lets apply OLS
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red')
```

```python
# Lets apply Gradient Descent assuming slope is constant m = 78.35
# and let's assume the starting value for intercept b = 100
y_pred = ((78.35 * X) + 100).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='#00a65a',label='b = 100')
plt.legend()
plt.show()
```

```python
m = 78.35
b_1 = 100

loss_slope_1 = -2 * np.sum(y - m*X.ravel() - b_1)
loss_slope_1
```

```python
# Lets take learning rate = 0.1
lr = 0.1

step_size = lr * loss_slope_1
step_size
```

```python
# Calculating the new intercept
b_2 = b_1 - step_size
b_2
```

```python
y_pred1 = ((78.35 * X) + b_2).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred1,color='blue',label='b = {}'.format(round(b_2,2)))
plt.plot(X,y_pred,color='green',label='b = 100')
plt.legend()
plt.show()
```

```python
# Iteration 2
loss_slope_2 = -2 * np.sum(y - m*X.ravel() - b_2)
loss_slope_2
```

```python
step_size = lr * loss_slope_2
step_size
```

```python
b_3 = b_2 - step_size
b_3
```

```python
y_pred2 = ((78.35 * X) + b_3).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred2,color='blue',label='b = {}'.format(round(b_3,2)))
plt.plot(X,y_pred1,color='green',label='b = {}'.format(round(b_2,2)))
plt.plot(X,y_pred,color='orange',label='b = 100')
plt.legend()
plt.show()
```

```python
# Iteration 3
loss_slope_3 = -2 * np.sum(y - m*X.ravel() - b_3)
loss_slope_3
```

```python
step_size = lr * loss_slope_3
step_size
```

```python
b_4 = b_3 - step_size
b_4
```

```python
y_pred3 = ((78.35 * X) + b_4).reshape(4)

plt.figure(figsize=(10,8))
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred3,color='black',label='b = {}'.format(round(b_4,2)))
plt.plot(X,y_pred2,color='blue',label='b = {}'.format(round(b_3,2)))
plt.plot(X,y_pred1,color='green',label='b = {}'.format(round(b_2,2)))
plt.plot(X,y_pred,color='orange',label='b = 100')
plt.legend()
plt.show()
```

```python
# now using loop for all iterations and setting value for b = -100
b = -100
m = 78.35
lr = 0.01

epochs = 50

for i in range(epochs):
  loss_slope = -2 * np.sum(y - m*X.ravel() - b)
  b = b - (lr * loss_slope)

  y_pred = m * X + b

  plt.plot(X,y_pred)

plt.scatter(X,y)
```
